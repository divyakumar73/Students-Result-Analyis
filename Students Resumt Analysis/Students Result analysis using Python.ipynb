# Drop Unnamed column

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
df.head()
```

# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

from the above chart we have analyzed that the number of female students are more than male students

# 

# Effect of parents education on the students score

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore": 'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb ,annot = True)
plt.title("Relationship between Parent`s Education and Student`s Score")
plt.show()
```

from the above chart we have concluded the education of the  parents have good impact on students marks

# Parent Marital score

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore": 'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1 ,annot = True)
plt.title("Relationship between Parent`s Marital Status and Student`s Score")
plt.show()
```

from the above chart we have concluded the Marital Status of the  parents have negligible impact on students marks
